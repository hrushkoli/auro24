```python
import numpy as np
import pandas as pd 
import re
import nltk
import spacy
import string

df = pd.read_csv('./review.csv')
df.head()
```

```python
df.rename(columns={"Review Text":'text'},inplace=True)

df.head()
```

```python
df["text_lower"] = df["text"].str.lower()
df[['text','text_lower']].head()
```

```python
PUNCT_TO_REMOVE = string.punctuation
print(PUNCT_TO_REMOVE)
```

```python
def remove_punctuation(text):
    return str(text).translate(str.maketrans('','',PUNCT_TO_REMOVE))

df["text_rv_punct"] = df["text"].apply(lambda text: remove_punctuation(text))
df[['text','text_rv_punct']].head()
```

```python
def remove_round_brackets(text):
    return re.sub('\(.*?\)','')
```

```python
df["text_rv_brackets"] = df["text"].apply(lambda text: remove_round_brackets(text))
df[['text','text_rv_brackets']].head()
```
